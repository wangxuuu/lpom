```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_acc_mnist_sgd = np.load("train_acc_mnist_sgd.npy" )
test_acc_mnist_sgd = np.load("test_acc_mnist_sgd.npy")

train_acc_mnist_lpom = np.load("train_acc_mnist_lpom.npy")
test_acc_mnist_lpom = np.load("test_acc_mnist_lpom.npy")

train_acc_mnist_cifar = np.load("train_acc_mnist_cifar.npy")
test_acc_mnist_cifar = np.load("test_acc_mnist_cifar.npy")

train_acc_cifar_lpom = np.load("train_acc_cifar_lpom.npy")
test_acc_cifar_lpom = np.load("test_acc_cifar_lpom.npy")
```

```python
plt.figure()
plt.plot(train_acc_mnist_sgd, label='SGD')
plt.plot(train_acc_mnist_lpom, label='LPOM')
plt.title('Train accuracy in MNIST')
plt.legend()
plt.savefig('./Train accuracy in MNIST.png')
plt.show()
```

```python
plt.figure()
plt.plot(test_acc_mnist_sgd, label='SGD')
plt.plot(test_acc_mnist_lpom, label='LPOM')
plt.title('Test accuracy in MNIST')
plt.legend()
plt.savefig('./Test accuracy in MNIST.png')
plt.show()
```

```python
plt.figure()
plt.plot(train_acc_mnist_cifar[:60], label='SGD')
plt.plot(train_acc_cifar_lpom, label='LPOM')
plt.title('Train accuracy in cifar')
plt.legend()
plt.savefig('./Train accuracy in cifar.png')
plt.show()
```

```python
plt.figure()
plt.plot(test_acc_mnist_cifar[:60], label='SGD')
plt.plot(test_acc_cifar_lpom, label='LPOM')
plt.title('Test accuracy in cifar')
plt.legend()
plt.savefig('./Test accuracy in cifar.png')
plt.show()
```
